```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd=pd.read_csv('gas_prices.csv')
pd.head()
```

```python
plt.figure(figsize=(12,6))

plt.title('Gas Prices over Time (in USD)')
plt.plot(pd.Year,pd.USA,'r.-')
plt.plot(pd.Year,pd.Canada,'b.-')
plt.plot(pd.Year,pd['South Korea'],'g.-')
plt.xticks(pd.Year[::3])
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()

plt.show()
```

```python
## tüm ülkeleri göstermek için böyle bir döngü yazılmalıdır.


plt.figure(figsize=(12,6))

plt.title('Gas Prices over Time (in USD)')
plt.xticks(pd.Year[::3])
for i in pd:
  if i != 'Year':
    plt.plot(pd.Year,pd[i],marker='.')
    print(i)
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend(loc='upper left')
plt.show()
```

```python
##Eğer belirli ülkeleri göstermek istersek
plt.figure(figsize=(12,6))

plt.title('Gas Prices over Time (in USD)')
plt.xticks(pd.Year[::3])
special_country=['France','USA','Germany','Italy']
for i in pd:
  if i in special_country:
    plt.plot(pd.Year,pd[i],marker='.')
    print(i)
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend(special_country,loc='upper left')
plt.savefig('Gasprice_figure',dpi=300)
plt.show()
```
